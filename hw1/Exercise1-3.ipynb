```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister

qr = QuantumRegister(10,'qr')

even = ClassicalRegister(5,'even')
odd = ClassicalRegister(5,'odd')

qc = QuantumCircuit(qr, even, odd)

for qri in range(10):
    if qri%2 == 0:
        qc.measure([qr[qri]], [even[qri//2]])
    else:
        qc.measure([qr[qri]], [odd[qri//2]])

qc.draw('mpl')
```
