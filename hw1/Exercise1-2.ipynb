```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister

qrx = QuantumRegister(1,'qx')
qry = QuantumRegister(1,'qy')
qrz = QuantumRegister(1,'qz')

crx = ClassicalRegister(1,'cx')
cry = ClassicalRegister(1,'cy')
crz = ClassicalRegister(1,'cz')

qc = QuantumCircuit(qrx,qry,qrz,crx,cry,crz)
qc.measure([qrx[0],qry[0],qrz[0]], [crx[0],cry[0],crz[0]])
qc.draw('mpl')
```
